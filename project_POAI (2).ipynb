# Splitting the Data:

```python
Ntest = 6
train = df_fc.iloc[:-Ntest]
test = df_fc.iloc[-Ntest:]

train_idx = df_fc.index <= train.index[-1]
test_idx = df_fc.index > train.index[-1]

#Defining a Plotting Function: 

def plot_fit_and_forecast(result, df_col, name):
    fig, ax = plt.subplots(figsize = (10,4))
    ax.plot(df_col, label = 'data')
    #plot the fitted curve on the train set
    train_pred = result.fittedvalues
    ax.plot(train.index, train_pred, color = 'green', label = 'in_sample fitted')
    
    #forecast the test set
    pred_result = result.get_forecast(Ntest)
    
    #calling conf_int() produces a dataframe with 2 columns: lower & upper values
    conf_int = pred_result.conf_int()
    lower, upper = conf_int['lower '+ name], conf_int['upper '+ name]
    forecast = pred_result.predicted_mean
    
    #plotting the forecasted values
    ax.plot(test.index, forecast, label = 'forecast')
    ax.set_xticklabels(labels = np.arange(1991, 2022, 1), rotation = 90)
    
    #create a shaded area in the plot with the confidence interval (alpha = transparency)
    ax.fill_between(test.index, lower, upper, color = 'blue', alpha = 0.2)
    ax.set_title('Unemployment forecast')
    ax.legend()
```

# Performing Augmented Dickey-Fuller Test:



```python

from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
def adf(x):
    res = adfuller(x)
    print("Test-statistic:", res[0])
    print("p-value:", res[1])
    if res[1] <0.05:
        print('Stationary')
    else:
        print('Non-stationary')
        
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt

years = np.arange(1991, 2022)
gdp_per_capita_India=df["India"].tolist()
gdp_per_capita_Russian_Federation=df["Russian Federation"].tolist()
gdp_per_capita_United_States=df["United States"].tolist()
gdp_per_capita_Europe_ca=df["Europe & Central Asia"].tolist()
gdp_per_capita_Ireland=df["Ireland"].tolist()

# Create a DataFrame
data_india = pd.DataFrame(data={'Year': years, 'GDP_Per_Capita': gdp_per_capita_India})
data_Russian_Federation = pd.DataFrame(data={'Year': years, 'GDP_Per_Capita': gdp_per_capita_Russian_Federation})
data_United_States = pd.DataFrame(data={'Year': years, 'GDP_Per_Capita': gdp_per_capita_United_States})
data_Europe_ca = pd.DataFrame(data={'Year': years, 'GDP_Per_Capita': gdp_per_capita_Europe_ca})
data_Ireland = pd.DataFrame(data={'Year': years, 'GDP_Per_Capita': gdp_per_capita_Ireland})


future_years = np.arange(2021, 2031)
# Fit ARIMA model
#India
model_india = ARIMA(data_india['GDP_Per_Capita'], order=(1, 1, 1))  # You may need to adjust order based on your data
result_india = model_india.fit()
forecast_india = result_india.get_forecast(steps=len(future_years))# Forecast future values
confidence_intervals_india = forecast_india.conf_int()

#Russian Federation
model_Russian_Federation = ARIMA(data_Russian_Federation['GDP_Per_Capita'], order=(1, 1, 1))  # You may need to adjust order based on your data
result_Russian_Federation = model_Russian_Federation.fit()
forecast_Russian_Federation = result_Russian_Federation.get_forecast(steps=len(future_years))# Forecast future values
confidence_intervals_Russian_Federation = forecast_Russian_Federation.conf_int()

#United States
model_United_States = ARIMA(data_United_States['GDP_Per_Capita'], order=(1, 1, 1))  # You may need to adjust order based on your data
result_United_States = model_United_States.fit()
forecast_United_States = result_United_States.get_forecast(steps=len(future_years))# Forecast future values
confidence_intervals_United_States = forecast_United_States.conf_int()

#Europe and Central Asia
model_Europe_ca = ARIMA(data_Europe_ca['GDP_Per_Capita'], order=(1, 1, 1))  # You may need to adjust order based on your data
result_Europe_ca = model_Europe_ca.fit()
forecast_Europe_ca = result_Europe_ca.get_forecast(steps=len(future_years))# Forecast future values
confidence_intervals_Europe_ca = forecast_Europe_ca.conf_int()

#Ireland
model_Ireland = ARIMA(data_Ireland['GDP_Per_Capita'], order=(1, 1, 1))  # You may need to adjust order based on your data
result_Ireland = model_Ireland.fit()
forecast_Ireland = result_Ireland.get_forecast(steps=len(future_years))# Forecast future values
confidence_intervals_Ireland = forecast_Ireland.conf_int()

# Plot the results
plt.figure(figsize=(20,20))
plt.subplot(3, 2, 1)
plt.plot(data_india['Year'], data_india['GDP_Per_Capita'], marker='o', linestyle='-', color='b', label='Historical Data')
plt.plot(future_years, forecast_india.predicted_mean, color='r', label='Forecasted Values')
plt.fill_between(future_years, confidence_intervals_india['lower GDP_Per_Capita'], confidence_intervals_india['upper GDP_Per_Capita'], color='pink', alpha=0.3, label='Confidence Intervals')
plt.xlabel('Year')
plt.ylabel('GDP Per Capita')
plt.title('Forecast of India unemployement')
plt.legend(loc = 'upper left')

plt.subplot(3, 2, 2)
plt.plot(data_Russian_Federation['Year'], data_Russian_Federation['GDP_Per_Capita'], marker='o', linestyle='-', color='b', label='Historical Data')
plt.plot(future_years, forecast_Russian_Federation.predicted_mean, color='r', label='Forecasted Values')
plt.fill_between(future_years, confidence_intervals_Russian_Federation['lower GDP_Per_Capita'], confidence_intervals_Russian_Federation['upper GDP_Per_Capita'], color='pink', alpha=0.3, label='Confidence Intervals')
plt.xlabel('Year')
plt.ylabel('unemployement')
plt.title('Forecast of Russian Federation unemployement')
plt.legend(loc = 'upper left')

plt.subplot(3, 2, 3)
plt.plot(data_United_States['Year'], data_United_States['GDP_Per_Capita'], marker='o', linestyle='-', color='b', label='Historical Data')
plt.plot(future_years, forecast_United_States.predicted_mean, color='r', label='Forecasted Values')
plt.fill_between(future_years, confidence_intervals_United_States['lower GDP_Per_Capita'], confidence_intervals_United_States['upper GDP_Per_Capita'], color='pink', alpha=0.3, label='Confidence Intervals')
plt.xlabel('Year')
plt.ylabel('unemployement')
plt.title('Forecast of United States unemployement')
plt.legend(loc = 'upper left')

plt.subplot(3, 2, 4)
plt.plot(data_Europe_ca['Year'], data_Europe_ca['GDP_Per_Capita'], marker='o', linestyle='-', color='b', label='Historical Data')
plt.plot(future_years, forecast_Europe_ca.predicted_mean, color='r', label='Forecasted Values')
plt.fill_between(future_years, confidence_intervals_Europe_ca['lower GDP_Per_Capita'], confidence_intervals_Europe_ca['upper GDP_Per_Capita'], color='pink', alpha=0.3, label='Confidence Intervals')
plt.xlabel('Year')
plt.ylabel('unemployement')
plt.title('Forecast of Europe & Central Asia unemployement')
plt.legend(loc = 'upper left')

plt.subplot(3, 2, 5)
plt.plot(data_Ireland['Year'], data_Ireland['GDP_Per_Capita'], marker='o', linestyle='-', color='b', label='Historical Data')
plt.plot(future_years, forecast_Ireland.predicted_mean, color='r', label='Forecasted Values')
plt.fill_between(future_years, confidence_intervals_Ireland['lower GDP_Per_Capita'], confidence_intervals_Ireland['upper GDP_Per_Capita'], color='pink', alpha=0.3, label='Confidence Intervals')
plt.xlabel('Year')
plt.ylabel('GDP Per Capita')
plt.title('Forecast of Ireland unemployement')
plt.legend(loc = 'upper left')

plt.show()


data_india = pd.DataFrame(data={'Year': years, 'GDP_Per_Capita': gdp_per_capita_India})
data_Russian_Federation = pd.DataFrame(data={'Year': years, 'GDP_Per_Capita': gdp_per_capita_Russian_Federation})
data_United_States = pd.DataFrame(data={'Year': years, 'GDP_Per_Capita': gdp_per_capita_United_States})
data_Europe_ca = pd.DataFrame(data={'Year': years, 'GDP_Per_Capita': gdp_per_capita_Europe_ca})


# Evaluate the model
mse_india = mean_squared_error(data_india['GDP_Per_Capita'], result_india.fittedvalues)
rmse_india = sqrt(mse_india)
print(f'Root Mean Squared Error of india model: {rmse_india}')

mse_Russian_Federation = mean_squared_error(data_Russian_Federation['GDP_Per_Capita'], result_Russian_Federation.fittedvalues)
rmse_Russian_Federation = sqrt(mse_Russian_Federation)
print(f'Root Mean Squared Error of Russian Federation model: {rmse_Russian_Federation}')

mse_United_States = mean_squared_error(data_United_States['GDP_Per_Capita'], result_United_States.fittedvalues)
rmse_United_States = sqrt(mse_United_States)
print(f'Root Mean Squared Error of United States model: {rmse_United_States}')

mse_Europe_ca = mean_squared_error(data_Europe_ca['GDP_Per_Capita'], result_Europe_ca.fittedvalues)
rmse_Europe_ca = sqrt(mse_Europe_ca)
print(f'Root Mean Squared Error of Europe & Central Asia model: {rmse_Europe_ca}')

mse_Ireland = mean_squared_error(data_Ireland['GDP_Per_Capita'], result_Ireland.fittedvalues)
rmse_Ireland = sqrt(mse_Ireland)
print(f'Root Mean Squared Error of Ireland model: {rmse_Ireland}')
```
